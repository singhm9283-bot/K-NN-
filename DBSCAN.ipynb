```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('circles_dataset.csv')
```

```python
df.info()
```

```python
sns.scatterplot(data=df, x='Feature 1',y='Feature 2')
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2, min_samples=3)
```

```python
dbs.fit(df)
```

```python
clusters=dbs.labels_
print(clusters)
```

```python
# -1 are the noise points
```

silhouette  score

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,clusters)
```
