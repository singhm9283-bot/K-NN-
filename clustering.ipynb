```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('student_clustering.csv')
```

```python
df.info()
```

```python
sns.scatterplot(data=df, x='cgpa',y='iq')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4) # 4 explains the no of clusters
```

```python
model.fit_predict(df)
```

INERTIA

```python
model.inertia_
```

```python
inr_list=[]
```

```python
for i in range(1,15):
  model= KMeans(n_clusters=i)
  model.fit_predict(df)
  inr_list.append(model.inertia_)
```

```python
print(inr_list)
```

### Elbow Method to find Optimal K

```python
plt.plot(range(1, 15), inr_list)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal K')
plt.show()
```
